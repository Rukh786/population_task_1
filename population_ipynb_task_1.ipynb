```python
import pandas as pd
import matplotlib.pyplot as plt

# Example Age Data
age = [15, 22, 25, 30, 35, 40, 45, 50, 22, 34, 30, 25, 19, 75, 80]

# Plot Histogram
plt.hist(age, bins=9, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
categories = ['Male', 'Female']
counts = [50, 60]

# Bar chart
plt.bar(categories, counts, color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
